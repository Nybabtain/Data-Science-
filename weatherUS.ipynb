# How did the snowfall & rainfall of United States changed over the years(1961 - 1990)

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
#Read the CSV file
df_sf = pd.read_csv('Desktop/UNdata.csv')
df_sf
```

```python
#Read the EXCEL file
df_rf = pd.read_excel('Desktop/Rain_1961_1990.xls')
df_rf.head()
```

```python
#Get the mean value of all the snowfall months through (1961-1999)
Jan = df_sf['Jan'].mean()
Feb = df_sf['Feb'].mean()
Mar = df_sf['Mar'].mean()
Apr = df_sf['Apr'].mean()
May = df_sf['May'].mean()
Jun = df_sf['Jun'].mean()
Jul = df_sf['Jul'].mean()
Aug = df_sf['Aug'].mean()
Sep = df_sf['Sep'].mean()
Oct = df_sf['Oct'].mean()
Nov = df_sf['Nov'].mean()
Dec = df_sf['Dec'].mean()
```

```python
#Creating a df to insert the values we've got above 
weatherOfUs = pd.DataFrame({'Snowfall rate':[Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov,Dec]})
#Changing the index names to months names
weatherOfUs.index= ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
weatherOfUs
```

```python
#Getting the mean values by grouping the months.
rainMonthly = df_rf.groupby(' Month').mean()['pr']
rainMonthly.iloc[0]
Jan = rainMonthly.iloc[0]
Feb = rainMonthly.iloc[1]
Mar = rainMonthly.iloc[2]
Apr = rainMonthly.iloc[3]
May = rainMonthly.iloc[4]
Jun = rainMonthly.iloc[5]
Jul = rainMonthly.iloc[6]
Aug = rainMonthly.iloc[7]
Sep = rainMonthly.iloc[8]
Oct = rainMonthly.iloc[9]
Nov = rainMonthly.iloc[10]
Dec = rainMonthly.iloc[11]
```

```python
#Inserting the rainfall into the df
weatherOfUs["Rainfall rate"] = [Jan,Feb,Mar,Apr,May,Jun,Jul,Aug,Sep,Oct,Nov,Dec]
weatherOfUs
```

```python
#We divide the snowfall rate by 1000 to get it like the rainfall rate.
weatherOfUs['Snowfall rate'] = weatherOfUs['Snowfall rate']/1000
weatherOfUs
```

```python

plt.plot(weatherOfUs.index, weatherOfUs['Snowfall rate'],label = "Snowfall Rate", alpha=0.5, marker= 'o')
plt.plot(weatherOfUs.index, weatherOfUs['Rainfall rate'],label = "Rainfall Rate", c='g', alpha = 0.5, marker = '*')

#Setting the y-axis to begin from 0.
plt.ylim(0)

#X-axis & Y-axis
plt.xlabel('Months of the year')
plt.ylabel('Rate of snow & rain')
#Title 
plt.title('Rates of Snowfall & Rain fall in the USA (1961-1990)')


plt.legend(frameon=False)

plt.show()
```
